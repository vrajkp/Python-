## Solution

1. fetch cnn site in Beautiful soap obejct
2. Get all the company names and tickers in a list
3. Pass company list to yahoo website and get details from the site and add everything into CSV
4. Ask user for input and search through the file for given ticker
5. Display details accordingly 

```python
import csv
from csv import writer
from bs4 import BeautifulSoup
import urllib.request, urllib.parse, urllib.error

#Function for Appending list as a row in the txt file
def add_row(file_name, elem):
# Open file in append mode
    with open(file_name, 'a+', newline='') as write_obj:
    # Create a writer object from csv module
      csv_writer = writer(write_obj)
# Add contents of list as last row in the csv file
      csv_writer.writerow(elem)
      write_obj.close()

site= urllib.request.urlopen('https://money.cnn.com/data/hotstocks/')
#print(site)
companies=[]
tickers=[]
soup=BeautifulSoup(site)
#print("SOUP: ",soup)
tables = soup.findAll("table", attrs={"class": "wsod_dataTable wsod_dataTableBigAlt"})
#print(tables[1])

for i in tables:
    # find name of the companies from the span tag
    for link in i.find_all("span"):
        #print(link.get("title"))
        # fetch name from the title if value exist in title tag
        if link.get("title")!=None:
            companies.append(link.get("title"))
    #print(companies)
    #fetch tickers of the cpmpany from anchor tags
    for  j in i.find_all("a"):
            tickers.append(j.text);

open('stocks.csv', 'w').close()

for i in range(len(companies)):
    
    elem=[]
    
    #adding symbol name into the website in order to get stock details of that specific ticker.
    yahoo_site="https://finance.yahoo.com/quote/"+tickers[i]+"?+p="+tickers[i]+"&.tsrc=fin-srch-v1"
    y_hand=urllib.request.urlopen(yahoo_site);
    y_soap=BeautifulSoup(y_hand)
    
    #finding prev close from website attribute value
    yaho_close=y_soap.find("td", attrs={"data-test":"PREV_CLOSE-value"})
    prev_close_value=yaho_close.text
    
    #finding open value from  webiste attribute 
    yaho_open=y_soap.find("td", attrs={"data-test":"OPEN-value"})
    open_val=yaho_open.text
    
    #finding volume value from  webiste attribute 
    yaho_vol=y_soap.find("td", attrs={"data-test":"TD_VOLUME-value"})
    vol_val=yaho_vol.text
    
    #finding market cam from  webiste attribute 
    yaho_market=y_soap.find("td", attrs={"data-test":"MARKET_CAP-value"})
    
    market_val=yaho_market.text
    if (i<=9):
        elem.append('Most Actives')        
        elem.append(tickers[i])    
        elem.append(companies[i])
        elem.append(float(open_val))
        elem.append(float(prev_close_value))
        elem.append(vol_val)
        elem.append(str(market_val))
    elif (i>9 and i<20):
        elem.append('Gainers')        
        elem.append(tickers[i])    
        elem.append(companies[i])
        elem.append(float(open_val))
        elem.append(float(prev_close_value))
        elem.append(vol_val)
        elem.append(str(market_val))
    else:
        elem.append('Loosers')        
        elem.append(tickers[i])    
        elem.append(companies[i])
        elem.append(float(open_val))
        elem.append(float(prev_close_value))
        elem.append(vol_val)
        elem.append(str(market_val))
        
#appending list of values into the txt file
    add_row('stocks.csv',elem)

print("This is a program to scrape data from the https://money.cnn.com/data/hotstocks/  for a class project. \nWhich stock are you interested in: ")

with open('stocks.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',' )
    #print("Most Actives:\n",reader)
    for i,row in enumerate(reader):
        #print(row[1])
        if(i == 10):
            print("\nGainers:\n")
        if(i==20):
            print("Loosers:\n")
        print(row[1],"",row[2]) 
        
flag = True
while(flag):

    with open('stocks.csv', newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=',' )
        key=(input("User inputs:"))
        if key not in tickers:
            print("Enter Valid ticker:")
        else:
            for i,row in enumerate(reader):
                if key in row:
                    print("\nThe data for","", row[1],"",row[2], "is the following:\n")
                    print(row[1]," ",row[2])
                    print("OPEN: ",row[3])
                    print("PREV CLOSE: ",row[4])
                    print("VOLUME: ",row[5])
                    print("MARKET CAP: ",row[6])
                    flag = False
                    flag1 = True
                    while(flag1):
                        play = input("Do you want to continue [yes or no]:")
                        if(play.lower() == "yes"):
                            break
                        elif(play.lower() == "no"):
                            flag1 = False
                            break
                        else:
                            print("Enter valid choice[yes or no]: ")
                    if(flag1 == False):
                        flag = False
                        break
                    else: 
                        break
```
